```python
import polars as pl
import glob
import os

# Folder and pattern
folder = 'gawnrb/meteo'
pattern = 'VRXZ07.*'

# Full search path
search_path = os.path.join(folder, pattern)

# List all matching files
file_paths = glob.glob(search_path)

# Read and collect dataframes
dfs = []
for file_path in file_paths:
    try:
        print(f"Reading: {file_path}")
        df = pl.read_csv(file_path, separator=';', skip_rows=2)
        dfs.append(df)
        print(df)
    except Exception as e:
        print(f"Failed to read {file_path}: {e}")
```

```python
dfs_casted = [df.cast(pl.String) for df in dfs]
combined_df = pl.concat(dfs_casted)

combined_df
```

```python
import polars as pl

# Step 1: Cast all input dfs to string for safe concat
dfs_casted = [df.cast(pl.String) for df in dfs]
combined_df = pl.concat(dfs_casted)

# Step 2: Identify float-able columns (skip metadata like "timestamp" or "id" if needed)
# For example, assuming you want to exclude 'timestamp' and 'source_file':
exclude_columns = {"timestamp", "source_file"}
float_columns = [col for col in combined_df.columns if col not in exclude_columns]

# Step 3: Convert these to float
combined_df = combined_df.with_columns([
    pl.col(col).cast(pl.Float64, strict=False) for col in float_columns
])

# Final result
print(combined_df)
```

```python
import polars as pl

# Refined headers
default_header = "Station/Location,Date,tre200s0,uor200s0,prestas0,fkl010z0,fkl010z1,dkl010z0,rre150z0"
second_header = "GAW-METEO,Date,temp,RH,ATPRES,WS1,WS2,WDIR,PRECIP"

# Paths
output_path = "D:/KADI-ASSIGNMENT/meteo.data.csv"
temp_data_path = "D:/KADI-ASSIGNMENT/meteo_data.csv"

# Overwrite first column with "GAW-METEO"
combined_df = combined_df.with_columns([
    pl.lit("GAW-METEO").alias(combined_df.columns[0])
])

# Save to temp file without header
combined_df.write_csv(temp_data_path, include_header=False)

# Write final CSV with refined headers
with open(output_path, 'w') as out_file:
    out_file.write(default_header + '\n')
    out_file.write(second_header + '\n')
    with open(temp_data_path, 'r') as temp_file:
        out_file.write(temp_file.read())


```
